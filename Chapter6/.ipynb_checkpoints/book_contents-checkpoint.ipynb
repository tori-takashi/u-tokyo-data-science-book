```python
data1 = {
    'id': ['100','101','102','103','104','106','108','110','111','113'],
    'city': ['Tokyo', 'Osaka', 'Kyoto', 'Hokkaido', 'Tokyo', 'Tokyo', 'Osaka', 'Kyoto', 'Hokkaido', 'Tokyo'],
    'birth_year': [1990, 1989, 1992, 1997, 1982, 1991, 1988, 1990, 1995, 1981],
    'name': ['Hiroshi', 'Akiko', 'Yuki', 'Satoru', 'Steeve', 'Mitsuru', 'Aoi', 'Tarou', 'Suguru', 'Mitsuo']}

df1 = DataFrame(data1)
df1
```

```python
data2 = {'id': ['100','101','102','105','107'],
        'math': [50,43,33,76,98],
        'english': [90,30,20,50,30],
        'sex':['M','F','F','M','M'],
        'index_num': [0,1,2,3,4]}
df2 = DataFrame(data2)
df2
```

```python
print('・結合テーブル：')
pd.merge(df1, df2, on='id')
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, left_index=True, right_on='index_num')
```

```python
pd.merge(df1, df2, how='left')
```

```python
city_map = {
    'Tokyo': 'Kanto',
    'Hokkaido': 'Hoddaido',
    'Osaka': 'Kansai',
    'Kyoto': 'Kansai'
}
city_map
```

```python
df1['region'] = df1['city'].map(city_map)
```

```python
df1
```

```python
df1['up_two_num'] = df1['birth_year'].map(lambda x: str(x)[0:3])
df1
```

```python
birth_year_bins = [1980, 1985, 1990, 1995, 2000]
birth_year_cut_data = pd.cut(df1.birth_year, birth_year_bins)
birth_year_cut_data
```

```python
pd.value_counts(birth_year_cut_data)
```

```python
groups = ['early1980s', 'late1980s', 'early1990s', 'late1990s']
birth_year_cut_data = pd.cut(df1.birth_year, birth_year_bins, labels=groups)
pd.value_counts(birth_year_cut_data)
```

```python
pd.cut(df1.birth_year, 2)
```

```python
pd.value_counts(pd.qcut(df1.birth_year, 3))
```
